Pregunta problema (Objetivo) 
-Que días de la semana se presentan más accidentes ?
-Cual es el actor vial más afectado ?
-Meses de menor accidentadilad respecto al año (2012-2023) ?

Contexto analitico - ** Fuente de datos **: [DATOS FUENTE](https://www.datos.gov.co/Transporte/Accidentes-de-Transito-ocurridos-en-el-Municipio-d/7cci-nqqb/about_data)

```python
C1 = A[['DÍA', 'DIURNIO/NOCTURNO']]

# Boxplot
plt.figure(figsize=(10, 6))
sns.boxplot(data=C1, x='DIURNIO/NOCTURNO', y='DÍA', showfliers=False, palette='pastel')

# Labels
plt.xlabel('D/N', fontsize=14)
plt.ylabel('Días', fontsize=14)
plt.title('Boxplot de Días por Tipo de Turno', fontsize=16, weight='bold')

# Layout
plt.subplots_adjust(left=0.1, right=0.9, bottom=0.1, top=0.9)

plt.show()
```

```python
# Prepare the data
C1 = A[['DÍA', 'GRAVEDAD', 'MES']]

# Create the boxplot
plt.figure(figsize=(12, 8))
sns.boxplot(data=C1, x='GRAVEDAD', y='DÍA', hue='MES', showfliers=False, palette='Set2')

# Set labels and title
plt.xlabel('Magnitud', fontsize=14)
plt.ylabel('Días', fontsize=14)
plt.title('Boxplot de Actores Viales por Gravedad y Mes', fontsize=16, weight='bold')

# Customize legend
plt.legend(title='', title_fontsize='13', fontsize='11', loc='upper center', bbox_to_anchor=(0.5, -0.1), ncol=4, frameon=False)

# Adjust layout for better spacing
plt.tight_layout()

# Display the plot
plt.show()
```

```python
conteo = A['GRAVEDAD'].value_counts()
print(conteo)
```

```python
C1 = A[['DÍA', 'GRAVEDAD', 'MES']]

plt.figure(figsize=(12, 6))
sns.histplot(C1['DÍA'], bins=20, color='skyblue', kde=True)

# Labels
plt.xlabel('Días', fontsize=14)
plt.ylabel('Frecuencia', fontsize=14)
plt.title('Distribución de Días', fontsize=16, weight='bold')

# Layout
plt.tight_layout()

plt.show()
```

```python
x = A['Propietario de Vehículo'].value_counts().index
y = A['Propietario de Vehículo'].value_counts().values

plt.figure(figsize=(12, 8))
sns.barplot(x=x, y=y, palette='viridis')

for i, value in enumerate(y):
    plt.text(i, value, f'{value}', ha='center', va='bottom', fontsize=12, color='black')

plt.xlabel('Propietario de Vehículo', fontsize=14)
plt.ylabel('Frecuencia', fontsize=14)
plt.title('Frecuencia por Clase', fontsize=16, weight='bold')

plt.xticks(rotation=45, ha='right', fontsize=12)

plt.tight_layout()

plt.show()
```

```python
# Create the catplot
g = sns.catplot(data=A, x="DÍA", kind="count", hue="GRAVEDAD", palette="Set2", height=6, aspect=2)

# Add labels and title
g.set_axis_labels("Día", "Conteo")
g.set_titles("Distribución de Días por Gravedad")
g.set_xticklabels(rotation=45, ha='right')
g.fig.suptitle("Conteo de Casos por Gravedad y Día", fontsize=16, weight='bold')

# Adjust layout to prevent clipping
plt.tight_layout()
plt.show()
```
