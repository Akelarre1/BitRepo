Pregunta problema (Objetivo) 
-Que días de la semana se presentan más accidentes ?
-Cual es el actor vial más afectado ?
-Meses de menor accidentadilad respecto al año (2012-2023) ?

Contexto analitico - ** Fuente de datos **: [DATOS FUENTE](https://www.datos.gov.co/Transporte/Accidentes-de-Transito-ocurridos-en-el-Municipio-d/7cci-nqqb/about_data)

```python
C1=A[['DÍA','DIURNIO/NOCTURNO']]
sns.boxplot(data=C1, x= 'DIURNIO/NOCTURNO', y='DÍA',showfliers= False)
plt.xlabel('D/N')
plt.ylabel('Días')
plt.title('Boxplot de Días')
plt.subplots_adjust(left=0.1, right=0.5, bottom=0.1, top=0.9, wspace=0.4, hspace=0.4)
```

```python
C1=A[['DÍA','GRAVEDAD','MES']]
sns.boxplot(data=C1, x= 'GRAVEDAD', y='DÍA',hue= 'MES',showfliers= False)
plt.xlabel('magnitud')
plt.ylabel('Días')
plt.title('Boxplot de Actores viales')
plt.subplots_adjust(left=0.1, right=2, bottom=0.1, top=2, wspace=2, hspace=15)
```
